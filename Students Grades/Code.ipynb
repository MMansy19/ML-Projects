# Lable Encoding (Find and replace)

```python
for col in df.columns:
  if type(df[col][0]) is str:
    print(col,df[col].unique())
```

```python
for col in df.columns:
  if df[col][0]=='yes' or df[col][0]=='no':
    df[col]=df[col].replace({'yes':'1','no':'0'},inplace=True)
df
```

# Splittung Data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
y=df['Average']
X=df.drop(['Average','G3'],axis=1)

X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
scalar=StandardScaler()
scalar
```

# Linear Regression

```python
# fit LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)

# predict
lr.score(X_test,y_test)

# Acuricy is so bad because of maney features
# so we will use lasso regression
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=1,max_iter=100,tol=0.1)
lasso.fit(X_train,y_train)
```
